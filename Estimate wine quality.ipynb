```python
#Calculating IQR Spread

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def outlier_treatment(datacolumn):
  
    #print(datacolumn)
    #sorted(datacolumn)
    Q1 , Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    
    return lower_range,upper_range,IQR
```
